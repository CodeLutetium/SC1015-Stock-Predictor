```python
import os
import pandas as pd
```

```python
dir_path = 'raw_aapl_tweets'
file_data = []
for file_name in os.scandir(dir_path):
    file_data.append(pd.DataFrame(pd.read_csv(f'{dir_path}/{file_name.name}')))
    
tweets_data = pd.concat(file_data).sort_values(by='created_at', ascending=False).reset_index().drop(labels='index', axis=1).drop_duplicates('full_text')
tweets_data = tweets_data[['created_at', 'full_text', 'favorite_count', 'view_count']]
tweets_data = tweets_data[tweets_data['full_text'].str.contains('apple', False) | tweets_data['full_text'].str.contains('aapl', False)]
tweets_data = tweets_data.reset_index().drop(labels='index', axis=1)
tweets_data
```

```python
tweets_data.to_csv('aapl_tweets.csv')
```
